```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
data=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
```

```python
data
```

```python
data['target']=diabetes.target
```

```python
data
```

```python
#data is already clean no extra EDA(DATA ANALYSIS) is required
```

```python
X= data[["bmi"]]  #one independent variable
Y=data["target"]  #dependent variable
```

```python
X
```

```python
Y
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y, test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# randomly spliting data default  if not want random use random_state=1,2,3,4,5,6,7,8,9
```

```python
# Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(x_train, y_train) # training of model
# at the backend it will use gradient descent to give optimal coefficient
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
# visualise
plt.scatter(x_test, y_test, label="actual data")
plt.plot(x_test, y_pred, color='red', linewidth=2, label="predicted data")
plt.legend()
plt.xlabel("BMI")
plt.ylabel("target")
plt.title("BMI vs target")
plt.show()
```

```python
# model evaluation
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
